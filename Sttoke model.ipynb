```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

le = LabelEncoder()

dataset2['gender'] = le.fit_transform(dataset2['gender'])
dataset2['ever_married'] = le.fit_transform(dataset2['ever_married'])
dataset2['work_type'] = le.fit_transform(dataset2['work_type'])
dataset2['Residence_type'] = le.fit_transform(dataset2['Residence_type'])
dataset2['smoking_status'] = le.fit_transform(dataset2['smoking_status'])

dataset2.head()
```

```python
dataset2['gender'] = pd.to_numeric(dataset2['gender'], errors='coerce')
dataset2['age'] = pd.to_numeric(dataset2['age'], errors='coerce')
dataset2['hypertension'] = pd.to_numeric(dataset2['hypertension'], errors='coerce')
dataset2['heart_disease'] = pd.to_numeric(dataset2['heart_disease'], errors='coerce')
dataset2['ever_married'] = pd.to_numeric(dataset2['ever_married'], errors='coerce')
dataset2['work_type'] = pd.to_numeric(dataset2['work_type'], errors='coerce')
dataset2['Residence_type'] = pd.to_numeric(dataset2['Residence_type'], errors='coerce')
dataset2['avg_glucose_level'] = pd.to_numeric(dataset2['avg_glucose_level'], errors='coerce')
dataset2['bmi'] = pd.to_numeric(dataset2['bmi'], errors='coerce')
dataset2['smoking_status'] = pd.to_numeric(dataset2['smoking_status'], errors='coerce')
```
